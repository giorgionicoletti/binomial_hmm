```python
import sys
sys.path.append('../')
```

```python
import numpy as np
import matplotlib.pyplot as plt

from lib.hmm_wrapper import binomial_HMM
```

```python
data = np.load('data.npy')
stateseq = np.load('stateseq.npy')
```

```python
D = data.shape[1]
N = 2
```

```python
hmm = binomial_HMM(N, D)
```

```python
print('Transition matrix')
print(np.exp(hmm.log_transition_matrix))
print()
print('Emission matrix')
print(np.exp(hmm.log_emission_matrix))
print()
print('Prior')
print(np.exp(hmm.log_state_priors))
```

```python
for _ in range(10):
    hmm.update(data)
    print()
    print('Transition matrix')
    print(np.exp(hmm.log_transition_matrix))
    print()
    print('Emission matrix')
    print(np.exp(hmm.log_emission_matrix))
    print()
    print('Prior')
    print(np.exp(hmm.log_state_priors))
    print()
```

```python
p = hmm.states_probability
```

```python
fig, ax = plt.subplots(figsize=(16,5))
xmin = 300
xmax = 650

plt.plot(np.arange(xmin, xmax, 1), p[:,1][xmin:xmax], label = 'Probability of inactive state', color = 'steelblue')
plt.plot(np.arange(xmin, xmax, 1), p[:,0][xmin:xmax], label = 'Probability of active state', color = 'coral')

ch = np.concatenate([np.array([xmin]),xmin+np.where(np.diff(stateseq[xmin:xmax]) != 0)[0],
                     xmin + np.array([len(stateseq[xmin:xmax])])])

colors = ['steelblue', 'coral']

for idx, i in enumerate(ch[:-1]):
    col = colors[stateseq[i+1]]
    plt.axvspan(i, ch[idx+1], facecolor=col, alpha=0.2)

plt.imshow(data[xmin:xmax].T, aspect = 'auto', cmap = 'Greys', interpolation = 'none', extent = [xmin,xmax,0,1])
plt.xlim(xmin,xmax)
plt.ylim(0,1.01)
plt.ylabel('Probability')
plt.legend()
plt.show()
```
